```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
```

```python
class TSDataset(Dataset):
    def __init__(self, df: pd.DataFrame, scaler, window_size: int = 5):
        self._window_size = window_size
        self._data = df
        self._scaler = scaler
    
    def __len__(self):
        return len(self._data)

    def __getitem__(self, idx: int):
        dataframe = self._data[idx: idx + self._window_size]
        y = self._data['KPI данные понедельно АлфаРМ_Продажи, рубли'].iloc[idx]
        return dataframe, y
    
```

```python
d = TSDataset(df, scaler)
```

```python
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot
# define dataset
X, y = df[df.columns[6:]][:244], df['KPI данные понедельно АлфаРМ_Продажи, рубли'][:244]

model = DecisionTreeRegressor()
model.fit(X, y)

importance = model.feature_importances_


k = sorted([(df.columns[i + 6], v) for i,v in enumerate(importance)], key=lambda x: x[1], reverse=True)
print(k[:5])

pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```
